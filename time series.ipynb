# Level 3 – Time Series Analysis

## Objective
To analyze trends and patterns in house prices using time series techniques.


```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv("raw_dataset.csv", sep=r"\s+", header=None)

df.columns = [
    "CRIM","ZN","INDUS","CHAS","NOX","RM","AGE",
    "DIS","RAD","TAX","PTRATIO","B","LSTAT","PRICE"
]

df['Time'] = range(len(df))
df.set_index('Time', inplace=True)

df.head()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['PRICE'])
plt.title("House Price Trend")
plt.xlabel("Time")
plt.ylabel("Price")
plt.show()
```

```python
decomposition = seasonal_decompose(df['PRICE'], model='additive', period=12)
decomposition.plot()
plt.show()
```

## Conclusion
Time series analysis helped identify trends and patterns in house prices.
Seasonal decomposition provided insight into underlying components.
